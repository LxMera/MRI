```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
epi_img = nib.load('sub-14:anat:sub-14_T1w.nii.gz')
epi_img_data = epi_img.get_fdata()
x,y,z=epi_img_data.shape

plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
plt.imshow(epi_img_data[int(x/2),:,:])
plt.subplot(1,3,2)
plt.imshow(epi_img_data[:,int(y/2),:])
plt.subplot(1,3,3)
plt.imshow(epi_img_data[:,:,int(z/2)])

np.set_printoptions(precision=2, suppress=True)
a1=epi_img.affine
a1
```

```python
np.linalg.inv(a1)
```
